### Задание 1. 
#### Базовое изучение
Изучить представленный набор данных на основе описания его столбцов и выбрать 8 столбцов для дальнейшего изучения (среди них должны быть как числовые, так и категориальные). Провести расчет базовых метрик для них, кратко описать результаты.


```python
import pandas as pd
file = 'https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/statistics_basics/horse_data.csv'

# прочитаем датасет
df = pd.read_csv(file, delimiter=',', na_values = '?')
# добавим колонки, чтобы датасет выглядел читабельнее
column_names = ['surgery', 'age', 'hospital_number', 'rectal_temperature','pulse', 'respiratory_rate',
                'temperature_of_extremities', 'peripheral_pulse', 'mucous_membranes', 'capillary_refill_time',
                'pain','peristalsis', 'abdominal_distension','nasogastric_tube','nasogastric_reflux',
                'nasogastric_reflux_PH','rectal_examination','abdomen','packed_cell_volume','total_protein',
                'abdominocentesis_appearance','abdomcentesis_total_protein','outcome','surgical_lesion',
                'type_of_lesion_1','type_of_lesion_2','type_of_lesion_3','cp_data']
df = pd.DataFrame(df.values, columns = column_names )
# посмотрим основную информацию по датасету
df.info()
# проверим установку заголовков
df.head()
```

```python
"""
Выберем данные, с которыми будем работать
Категориальные значения:
- surgery (наличие операции)
- age (возраст)
- temperature of extremities (температура конечностей)
- abdominal distension (растяжение брюшной полости)
- outcome (итоги, что в конце концов случилось с лошадью)
- peristalsis (перистальтика)
Числовые значения
- rectal_temperature (ректальная температура)
- pulse (пульс)

"""
df = df[['age', 'surgery', 'temperature_of_extremities', 'abdominal_distension',
         'peristalsis', 'outcome', 'rectal_temperature', 'pulse']]
# посмотрим основную информацию по датасету
df.info()
df.head()
```

```python
# Считаем базовые метрики
df.describe()
# получаем незначительный разброс по температуре и большой разброс по пульсу
```

### Задание 2.
#### Работа с выбросами
В выбранных числовых столбцах найти выбросы, выдвинуть гипотезы об их причинах и проинтерпретировать результаты. Принять и обосновать решение о дальнейшей работе с ними.

```python
# посмотрим выбросы по температуре
q1 = df['rectal_temperature'].quantile(0.25)
q3 = df['rectal_temperature'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['rectal_temperature'].between(lower_bound, upper_bound, inclusive=True)].sort_values('rectal_temperature')
pd.concat([df, remove_outliers]).drop_duplicates(keep=False)
# основные выбросы - незаполненные значения
```

```python
# заполним пустые значения медианой
df['rectal_temperature'].fillna(df['rectal_temperature'].median(), inplace=True)
df
```

```python
# посмотрим выбросы по пульсу
q1 = df['pulse'].quantile(0.25)
q3 = df['pulse'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['pulse'].between(lower_bound, upper_bound, inclusive=True)].sort_values('pulse')
pd.concat([df, remove_outliers]).drop_duplicates(keep=False)
# основные выбросы - незаполненные значения
```

```python
# заполним пустые значения медианой
df['pulse'].fillna(df['pulse'].median(), inplace=True)
df
```

### Задание 3.
#### Работа с пропусками
Рассчитать количество выбросов для всех выбранных столбцов. Принять и обосновать решение о методе работы с пропусками по каждому столбцу, сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
# посмотрим выбросы по возрасту
q1 = df['age'].quantile(0.25)
q3 = df['age'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['age'].between(lower_bound, upper_bound, inclusive=True)].sort_values('age')
pd.concat([df, remove_outliers]).drop_duplicates(keep=False)
# похоже на ошибку ввода данных, т.к. возраст обозначается в пределах 1-2
```

```python
# исправим 9 на 2
df.loc[(df['age'] == 9), 'age'] = 2
df
```

```python
# посмотрим выбросы по операциям
q1 = df['surgery'].quantile(0.25)
q3 = df['surgery'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - (1.5 * iqr) 
upper_bound = q3 + (1.5 * iqr)
remove_outliers = df[df['surgery'].between(lower_bound, upper_bound, inclusive=True)].sort_values('surgery')
pd.concat([df, remove_outliers]).drop_duplicates(keep=False)
# похоже на единичный пропуск
```

```python
# исправим на моду
df['surgery'].fillna(df['surgery'].mode()[0],inplace=True)
```
